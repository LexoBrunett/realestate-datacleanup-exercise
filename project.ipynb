#### Exercise 14. Now let's work with the "south belt" of madrid. Make a subset of the original DataFrame that contains the following populations (level5 column): "Fuenlabrada","Leganés","Getafe","Alcorcón" (★★☆)

Hint: Filter the original DataFrame using the column `level5` and the function `isin`

```python
# Exercise 14

#Get new dataset with the filtered values
south_belt_populations = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]
south_belt_subset = ds[ds['level5'].isin(south_belt_populations)]

south_belt_subset

#Replace commas to convert from string to float
south_belt_subset['latitude'] = south_belt_subset['latitude'].str.replace(',', '.')
south_belt_subset['longitude'] = south_belt_subset['longitude'].str.replace(',', '.')

# Convert values to float
south_belt_subset['latitude'] = south_belt_subset['latitude'].astype(float)
south_belt_subset['longitude'] = south_belt_subset['longitude'].astype(float)
```

#### Exercise 15. Make a bar plot of the median of the prices and explain what you observe (you must use the subset obtained in the question 14) (★★★)

Print the bar of the median of the prices and write in the Markdown cell a brief analysis about the plot

```python
# Calcular la mediana de los precios para cada población
median_prices = south_belt_subset.groupby('level5')['price'].median().sort_values()

# Plotear el gráfico de barras
median_prices.plot(kind='bar', color='orange')
plt.title('Median Prices by Population')
plt.xlabel('Population')
plt.ylabel('Median Price (USD)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# Importar la función de normalización de scikit-learn
from sklearn.preprocessing import MinMaxScaler

# Inicializar el escalador
scaler = MinMaxScaler()

# Normalizar los precios para cada población
normalized_prices = south_belt_subset.groupby('level5')['price'].apply(lambda x: scaler.fit_transform(x.values.reshape(-1, 1)).flatten())

# Plotear los histogramas normalizados
plt.figure(figsize=(10, 6))
for population, prices in normalized_prices.items():
    plt.hist(prices, bins=20, alpha=0.5, label=population)

plt.title('Normalized Prices Distribution by Population')
plt.xlabel('Normalized Price')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# Obtener las poblaciones a graficar
populations_to_plot = ["Fuenlabrada","Leganés","Getafe","Alcorcón"]

# Configurar los subgráficos
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
axes = axes.flatten()

# Hacer un gráfico de dispersión para cada población
for i, population in enumerate(populations_to_plot):
    subset = south_belt_subset[south_belt_subset['level5'] == population]
    axes[i].scatter(subset['surface'], subset['price'], alpha=0.5)
    axes[i].set_title(population)
    axes[i].set_xlabel('Surface')
    axes[i].set_ylabel('Price (USD)')
    axes[i].grid(True)

# Ajustar el diseño y mostrar el gráfico
plt.tight_layout()
plt.show()
```
